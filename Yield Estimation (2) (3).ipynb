# Using Label Encoding

```python
label_encode=preprocessing.LabelEncoder

def label_encode(dfn,column):
    label_encode=preprocessing.LabelEncoder()
    for col in column:
        dfn[col]=label_encode.fit_transform(dfn[col].astype(str))
    return dfn

columns_to_encode =['CCE_Id','Village','Farmer_Nam','Irrigation','Seed_Varity','Crop_Varie','Crop condition','Any_Damage','Weeds']

label_encoded=label_encode(dfn,columns_to_encode)
```
